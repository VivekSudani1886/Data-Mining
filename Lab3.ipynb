### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
quantitive = ['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']

for col in quantitive:
    print(col + ":")
    print("        Mean ",df[col].mean())
    print("        Standard deviation ",df[col].std())
    print("        Minimum ",df[col].min())
    print("        Maximum ",df[col].max())
    print("        Mode ",df[col].mode()[0])
    print("        Range ",df[col].max() - df[col].min())


```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
pairs = [
    ('Age', 'Fare'),
    ('Age', 'SibSp'),
    ('Age', 'Parch'),
    ('Age', 'SibSp'),
    ('Fare', 'SibSp'),
    ('Fare', 'Parch')
]

plt.figure(figsize=(15, 8))

for i, (x, y) in enumerate(pairs):
    plt.subplot(2, 3, i+1)
    plt.scatter(df[x], df[y], alpha=0.5, edgecolors='k')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f'{x} vs {y}')
    plt.grid(True)
    
plt.tight_layout()
plt.suptitle('Scatter Plots of Attribute Pairs', fontsize=16, y=1.02)
plt.show()
```

<h2>Slicling Excerice</h2>

```python
import numpy as py
arr = py.array([[101,102,103,104,105],[201,202,203,204,205],[301,302,303,304,305],[401,402,403,404,405]])
```

```python
#whole array
#odd roll no of b and d
#last roll no of all class
#only[202 to 204 and 302 to 304]
```

```python
#whole array
print(arr[::])
```

```python
#odd roll no of 2nd and 4th column

print(arr[1::2,::2])
```

```python
#last roll no of all class

print(arr[::,4::])
```

```python
#only[202 to 204 and 302 to 304]
print(arr[1:3:,1:4:])
```
